#### 加载停止词
* 停止词加载出现部分编码问题，因此部分停止词需要在代码中添加（带转义字符的），其他的都可以到停止词文件中添加
* 输入源停止词txt文件，如’stoplist.txt',去重后，返回列表格式的停止词

```python
infile = open('stoplist.txt', 'r', encoding='utf-16')
stopwordslist = []
stopwordslist.append("\u200b")
stopwordslist.append('\r\n')
stopwordslist.append('"')
for str in infile.read().split('\n'):
    if str not in stopwordslist:
        stopwordslist.append(str)
        #outfile.write(str + '\n')
# data = xlrd.open_workbook('stoplist.xlsx')
# table = data.sheets()[0]
# nrows = table.nrows
# ncols = table.ncols
# stopwordslist =table.col_values(0)

print ('停止词共有%d个'%len(stopwordslist) )
print ('最后一个停止词是',stopwordslist[0])
```

#### 清洗微博数据
* 非常简洁的代码
* 分词、删除停止词
* 返回一个清洗后的数据列表

```python
import jieba
texts = [[word for word in list(jieba.cut(sentence, cut_all = False)) if word not in stopwordslist]  
        for sentence in train] 
```

#### 使用gensim进行话题模型分析
* 输入的参数分别为：
* 清洗后的数据
* 主题数
* passes

```python
len(texts)
```

```python
from gensim import corpora  
import gensim
from gensim.models import LdaModel  
from gensim.corpora import Dictionary 

dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
Lda = gensim.models.ldamodel.LdaModel
ldamodel = Lda(corpus, num_topics = 10, id2word = dictionary, passes=20)
```

#### 打印最终结果

```python
print(ldamodel.print_topics(num_topics=3, num_words=20))
```

#### document_topics为输出的每一条微博的相应的主题的概率

```python
document_topics = [ldamodel[c] for c in corpus]         
```

#### 艰难但实现了8个列表

```python
topic0=[]
topic1=[]
topic2=[]
topic0_n=[]
topic1_n=[]
topic2_n=[]
topic_max=[]
topic_max_n=[]
for i in range(len(document_topics)):
    list1=len(document_topics[i])
    new0=[]
    for x in document_topics[i]:
        new0.append(int(x[0]))    
    if 0 in new0:
        m = new0.index(0)
        topic0.append('Yes')
        topic0_n.append(document_topics[i][m][1])
    else:
        topic0.append('NO')
        topic0_n.append(0)
    new1=[]
    for x in document_topics[i]:
        new1.append(int(x[0]))    
    if 1  in new1:
        m = new1.index(1)
        topic1.append('Yes')
        topic1_n.append(document_topics[i][m][1])
    else:
        topic1.append('NO')
        topic1_n.append(0)
    new2=[]
    for x in document_topics[i]:
        new2.append(int(x[0]))    
    if 2  in new2:
        m = new2.index(2)
        topic2.append('Yes')
        topic2_n.append(document_topics[i][m][1])
    else:
        topic2.append('NO')
        topic2_n.append(0)
    get_max=[]
    get_max.append(topic0_n[i])
    get_max.append(topic1_n[i])
    get_max.append(topic2_n[i])
    topic_max_n.append(max(get_max))
    index1=get_max.index(max(get_max))
    topic_max.append(index1)
```
